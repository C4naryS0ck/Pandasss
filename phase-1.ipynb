### Introduction to pandas Data Structures.
#### 1. Series :
is a one dimensional array-like object containing an array of data(of any Numpy data type) and an associated array of data labels, called its index. 

```python
obj = pd.Series([4,7,-5,3]) 
obj
# original index starting from 0 is shown
```

```python
print(obj.values)

print(obj.index)
```

##### Defining our own index :

```python
obj2 = pd.Series([4,5,7,-4], index = ['a','b','c','d'])

print(obj2)
print(obj2.index)
# you can use values in the index when selecting single values or a 
# set of values.
print(obj2['a'])
print(obj2[['d','c','a']])
```

### Operations

```python
print(obj2*2)
print('b' in obj2)
print('e' in obj2)
```

#### we can convert the data from python dictionary into Series.

```python
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = pd.Series(sdata)
print(obj3)
```

```python
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = pd.Series(sdata, index = states)
print(obj4)
```

#### Handling missing values:

```python
# the value is null
print(pd.isnull(obj4))
# the values is not null
print("\nnotnull vlaues\n",pd.notnull(obj4))

print("\ninstance methods\n",obj4.isnull())
```

## 2. DataFrame
It represents a tabular, spreadsheet-like data structure containg an ordered collection of columns, each of which can be a different value type(numeric, string, boolean, etc). 

#### How to construct a dataframe?
1. From a dict of equal-length lists or NumPy arrays.

```python
data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9]
}

frame = pd.DataFrame(data)
print(frame)

frame2 = pd.DataFrame(data, columns= ['year','state','pop'], 
                      index = ['one','two','three','four','five'])

print("\nframe2\n",frame2)

# passing a column that isn't contained in data it will appear NA.
frame3 = pd.DataFrame(data, columns= ['year','state','pop','debt'], 
                      index = ['one','two','three','four','five'])
print("\nFrame3\n", frame3)
```

#### Retreiving a column:

```python
# 1. BY DICT LIKE NOTATION:
print(frame2['state'])

# BY ATTRIBUTE: 
print("\nBy attribute:\n",frame2.state)
```

#### RETRIEVING A ROW

```python
# 1. POSITION : (.iloc)
print("\n",frame2)
print("\nBy Position: 1st row =>\n",frame2.iloc[0]) 
print("\nBy Position: 1st three rows =>\n",frame2.iloc[0:3]) 
print("\nBy Position: 3rd row, 2nd column =>\n",frame2.iloc[2,1]) 
```

```python
# BY LABEL/NAME : (.loc)

print("\nframe2:\n",frame2)
print("\nrow with label index three:\n", frame2.loc['three']) 
print("\nmultiple rows by label:\n", frame2.loc[['one','two','three']]) 
print("\nSpecific row and column by label:\n", frame2.loc['one','year']) 
```

#### Columns can be modified by assignment.

```python
print("frame3 :\n",frame3)
frame3['debt'] = 16.5

print("\nframe3 after assignment:\n",frame3) 

import numpy as np
frame3['debt'] = np.arange(5) 
print("\nframe3 after assignment with numpy:\n",frame3) 


```

```python
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'four', 'five'])  
frame2['debt'] = val
print("use of Nan in place of unfilled data:\n",frame2)

frame2['eastern'] = frame2.state == 'Ohio' 
print("\nAssigning a column that doesn't exist :\n",frame2) 

# DELETING A COLUMN:
del frame2['eastern']
print("\nDataFrame after deleted column :\n",frame2) 

```

### NESTED DICTIONARY

```python
pop = {
    'Nevada':{1:2.3,2:45},
    'Ohio' : {3:32,4:43}
}
frame4 = pd.DataFrame(pop)
print(frame4) 

print("\ntransposed: \n",frame4.T)
```
