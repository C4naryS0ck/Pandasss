### COPY VS VIEW

### COPY CHANGES ONLY THE COPY - ORIGINAL REMAINS THE SAME

```python
newdf2 = newdf.copy() # view is pointing, newdf2 is pointing to newdf. 
```

```python
newdf2.loc[0,0] = 1
```

```python
newdf2
```

```python
newdf 
```

### VIEW CHANGES BOTH THE ORIGINAL AND THE VIEW DATAFRAME.

```python
newdf2 = newdf
newdf2.loc[0,0] = 23
```

```python
newdf2
```

```python
newdf
```

```python
newdf.columns = list('abcde')
newdf.head(2)
```

```python
newdf.loc[0,0] = 69
newdf.head(2)
```

```python
newdf.loc[0,"a"] = 59
newdf.head(2)
```
