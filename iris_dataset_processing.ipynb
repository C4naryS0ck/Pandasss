## PROCESSING IRIS DATASET

```python
import pandas as pd
import numpy as np
```

### STEP 1: LOADING THE DATA

```python
df = pd.read_csv("iris.csv", encoding = 'latin1') 
print(df)

print("\nThe first 5 rows to see what we are working with:\n",df.head())
```

### STEP 2: EXPLORING THE DATA

```python
print("check how many rows and columns we have:\n",df.shape)

print("\nGet a quick summary of column,data types, and missing values:\n",df.info())

print("\nGet statistical summaries:\n",df.describe()) 
```

### STEP 3: HANDLING MISSING VALUES (Pandas & Numpy)

```python
print("Check for missing values in each column:\n")
print(df.isnull().sum())
```

### STEP 4: SPLITTING FEATURES AND LABELS (Pandas & Numpy)
In ML, you must separate your Features(the inpus:sepal length, sepal width, petal length, petal width) from you Labels (the target output: the species name).

```python
print("Dataset:\n", df)

# X = all columns except 'species' (features)
X = df.drop(columns=['species']).values 
# .values converts the Pandas DataFrame to a Numpy array.

# y = only the "species" column (lables)
y = df['species'].values

print("Features shape (Numpy array):", X.shape)
print("Labels shape (Numpy array):", y.shape) 
```

> Tip: Machine learning algorithms prefer NumPy arrays over Pandas DataFrames because NumPy operations are much faster and optimized for math.

### STEP 5: ENCODING THE TARGET VARIABLE (NumPy)

y variable contains strings like 'Iris-setosa','Iris-versicolor','Iris-virginica'. Computers only understand numbers, so we need to convert these text labels into 0,1 & 2.


```python
print("Find the unique species name and map them to numbers.")
unique_species, y_encoded = np.unique(y, return_inverse = True) 
print("Unique species found:",unique_species)
print("encoded labels as numbers:", y_encoded)
```
